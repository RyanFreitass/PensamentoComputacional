#2. Escreva um programa para uma urna eletrônica. O programa fica lendo os votos até que o usuário informe o valor 4444444 (código de finalização da urna). Após a finalização da urna, o programa deverá escrever na tela a quantidade de votos e o percentual de votos para cada uma das opções abaixo:

  a. Votos no Candidato 77

  b. Votos no Candidato 39

  c. Votos no Candidato 66

  d. Voto em branco (0)

  e. Votos nulos (nenhuma das opções anteriores)

  f. Votos válidos (aqueles computados para algum candidato)

  g. Votos não válidos

```python
from pydub import AudioSegment
from IPython.display import Audio, display

import time

# Caminho correto para o som dentro da pasta 'audio' no Google Drive
caminho_som = "/content/drive/MyDrive/audio/confirma-urna.mp3"

# Carrega o som da urna
urna_som = AudioSegment.from_mp3(caminho_som)

# Função para tocar o som
def tocar_som():
    urna_som.export("temp_som.mp3", format="mp3")
    display(Audio("temp_som.mp3", autoplay=True))
    time.sleep(2)


#ALGORITMO DOIS
voto = ""
contador1 = 0
contador2 = 0
contador3 = 0
contador4 = 0
p1 = 0.0
p2 = 0.0
p3 = 0.0
total_votos = 0
contador5 = 0

#Inicio
print("URNA ELETRÔNICA")
print("--"*10)

#candidatos
print("CANDIDATO 77")
print("CANDIDATO 39")
print("CANDIDATO 66")
print("--"*10)

voto = input("DIGITE O SEU VOTO NA URNA: ")

while True:
    if voto == "77":
        contador1 += 1
        tocar_som()
    elif voto == "39":
        contador2 += 1
        tocar_som()
    elif voto == "66":
        contador3 += 1
        tocar_som()
    elif voto == "0":
        contador4 += 1
        tocar_som()
    elif voto == "4444444":
        break
    else:
        contador5 += 1

    voto = input("DIGITE O SEU VOTO NA URNA: ")

#resultados urna
print("--"*10)
#VOTOS VALIDOS
print(f"VOTOS NO CANDIDATO 77: {contador1}")
print(f"VOTOS NO CANDIDATO 39: {contador2}")
print(f"VOTOS NO CANDIDATO 66: {contador3}")
print(f"VOTOS EM BRANCO [0]: {contador4}")
print(f"VOTOS VÁLIDOS: {contador1 + contador2 + contador3 + contador4}")

#VOTOS NAO VALIDOS
print(f"VOTOS NÃO VÁLIDOS: {contador4 + contador5}")

print("--"*10)

#percentual
total_votos = contador1 + contador2 + contador3
if total_votos > 0:
    p1 = (contador1 / total_votos) * 100
    p2 = (contador2 / total_votos) * 100
    p3 = (contador3 / total_votos) * 100

    print(f"Candidato 77: {p1:.2f}%")
    print(f"Candidato 39: {p2:.2f}%")
    print(f"Candidato 66: {p3:.2f}%")
else:
    print("Nenhum voto válido para calcular o percentual.")

print("--"*10)

print("FIM.")
```
